```python
# import
import torch
import torch.nn as nn
import torch.nn.functional as F

import torch.optim as optim
from torch.utils.data import DataLoader

import torchvision
import torchvision.transforms as transforms
```

```python
# gpu사용과 랜덤으로 만들때 고정된값을 사용하기위한 코드
device = 'cuda' if torch.cuda.is_available() else 'cpu'

torch.manual_seed(777)
if device =='cuda':
    torch.cuda.manual_seed_all(777)
```

```python
# To.Tensor()로 바꿔야 입력이 가능하기 때문에 train_data를 ImageFolder를 통해서 DataSet으로 사용하도록 바꿔줌
trans = transforms.Compose([
    transforms.ToTensor()
])

train_data = torchvision.datasets.ImageFolder(root='./custom_data/train_data', transform=trans)
```

```python
#DataLoader로 불러서 사용
data_loader = DataLoader(dataset = train_data, batch_size = 8, shuffle = True, num_workers=2)
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(3,6,5), #입력채널 3 출력채널 6 filtersize 5 stride와 padding은 default 가 각각 1 , 0이기떄문에 선언안함
            nn.ReLU(), 
            nn.MaxPool2d(2),
        )
        self.layer2 = nn.Sequential(
            nn.Conv2d(6,16,5), # 똫같이 채널사이즈만 바뀌어서 진행
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.layer3 = nn.Sequential(
            nn.Linear(16*13*29, 120), #fully connected두번진행
            nn.ReLU(),
            nn.Linear(120,2)
        )
        
    # print(out.shape)을 넣어줘서 아래 testing에 확인해보면 좋음
    def forward(self, x):
        out = self.layer1(x)
      
        out = self.layer2(out)
      
        out = out.view(out.shape[0], -1)
      
        out = self.layer3(out)
        return out
        
```

```python
#testing 
net = CNN().to(device) # 네트워크 생성
test_input = (torch.Tensor(3,3,64,128)).to(device) # test_input 생성
test_out = net(test_input) # 네트워크에 testinput 넣어줌
```

```python
optimizer = optim.Adam(net.parameters(), lr=0.00001) # Adam optimizer사용
loss_func = nn.CrossEntropyLoss().to(device) # loss 는 CrossEntropyLoss사용
```

```python
# dataSet이 쉬워서 잘학습되기 때문에 5번만 진행
total_batch = len(data_loader)

epochs = 5
for epoch in range(epochs):
    avg_cost = 0.0
    for num, data in enumerate(data_loader):
        imgs, labels = data
        imgs = imgs.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        out = net(imgs)
        loss = loss_func(out, labels)
        loss.backward()
        optimizer.step()
        
        avg_cost += loss / total_batch
        
    print('[Epoch:{}] cost = {}'.format(epoch+1, avg_cost))
print('Learning Finished!') 
```

```python
# 학습시킨 모델을 쓸때마다 학습할 필요는 없기 때문에 
# torch.save를 이용하여 그 네트워크에 state dict값을 저장
# 모델의 학습가능한 매개변수(가중치와 편향)들은 모델의 매개변수에 포함되어있다.
# 간단히 말해 state_dict는 각 계층을 매개변수 텐서로 매핑되는 Python 사전(dic)객체이다.
torch.save(net.state_dict(), "./model/model.pth")
```

```python
# 불러오려면 네트워크의 모형은 똫같애야하기 때문에 똫같은 모형으로 선언
new_net = CNN().to(device)
```
