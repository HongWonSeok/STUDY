**Multi Layer Perceptron**

image.png

Marvin Minsky라는 분이 하나의 layer를 가지는 perceptron구조론은 XOR를 해결 할 수 없다. 이 XOR문제를 해결하기 위해서는 Multi Layer Perceptron이라는게 필요하다라는 것을 증명했다.

Multi layer Perceptron은 말그대로 하나의 layer만 가지는 것이아니라 여러개의 layer를 가지는 perceptron을 말한다.
